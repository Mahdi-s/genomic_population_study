### This file is used to analyze and visulize the specific traits and associated assesments
Make sure you have final_df2.csv in the same folder as this file and run begin running each cell

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('final_df2.csv')
df
```

### Group by trait name
In this section we group our data by trait name

```python
newdf= df.groupby('trait name')
newdf.first()
```

### Pick specific trait name
Simply Insert a trait name into the cell below as indicated by the comment

```python
data=newdf.get_group('Fat Taste Perception')
data
```

```python
#drop train name from set for easier analysis
data = data.iloc[: , 1:]
data
```

```python
df11=data.groupby(['gender','ancestry'])['hypertention', 'bmi'].apply(lambda x: (x==1).sum())#.reset_index(name='count')
df11.plot(kind='barh',stacked=True,rot=1, title='Gender & Ancestry Segmentations for presence of hypertention & diabetes for slightly elevated assessment', figsize=(15,15))
plt.yticks(ticks = [0,1,2,3],labels = ['(M,AA)','(M,CAU)','(F,AA)','(F,CAU)'], rotation = 'horizontal')
plt.show()
```

```python
df11=data.groupby(['gender','ancestry'])['hypertention', 'bmi'].apply(lambda x: (x==0).sum())#.reset_index(name='count')
df11.plot(kind='barh',stacked=True,rot=1, title='Gender & Ancestry Segmentations for lack of hypertention & diabetes for slightly elevated assessment', figsize=(15,15))
plt.yticks(ticks = [0,1,2,3],labels = ['(M,AA)','(M,CAU)','(F,AA)','(F,CAU)'], rotation = 'horizontal')
plt.show()
```

```python
newdf2= df.groupby(['trait name', 'assesment'])
newdf2.first()
```

```python
data3=newdf2.get_group(('Difficulty in Losing Weight','SLIGHTLY ELEVATED'))
data3
```

```python
df11=data3.groupby(['gender','ancestry'])['hypertention', 'bmi'].apply(lambda x: (x==1).sum())#.reset_index(name='count')
df11.plot(kind='barh',stacked=True,rot=1, title='Gender & Ancestry Segmentations for presence of hypertention & diabetes for slightly elevated assessment', figsize=(15,15))
plt.yticks(ticks = [0,1,2,3],labels = ['(M,AA)','(M,CAU)','(F,AA)','(F,CAU)'], rotation = 'horizontal')
plt.show()
```
