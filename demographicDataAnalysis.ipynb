```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('demog_data.csv')
df
```

```python
df.describe()
```

```python
newdata=df.groupby(['gender']).size()
newdata
```

```python
# plot the result
newdata.plot(kind='pie', figsize=(20,10), legend=True, title='Population spread based on gender', autopct='%1.0f%%')
plt.xticks(rotation=90)
plt.show()
```

```python
newdata2=df.groupby(['ancestry']).size()
newdata2
```

```python
# plot the result
newdata2.plot(kind='pie', figsize=(20,10), legend=True, title='Population spread based on ancestry', autopct='%1.0f%%')
plt.xticks(rotation=90)
plt.show()
```

```python
newdata3=df.groupby(['ancestry', 'gender']).size()
newdata3
```

```python
# plot the result
newdata3.plot(kind='bar', figsize=(20,10), legend=True, title='Population spread based on gender & ancestry')
plt.xticks(rotation=30)
plt.show()
```

```python
newdata4=df.groupby(['bmi']).size()
newdata4
```

```python
# plot the result
newdata4.plot(kind='bar', figsize=(20,10), legend=True, title='Population BMI')
plt.xticks(rotation=30)
plt.show()
```
