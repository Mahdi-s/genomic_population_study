### This file is used to analyze and visulize the specific traits and associated assesments
Make sure you have final_df2.csv in the same folder as this file and run begin running each cell

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('final_df2.csv')
df
```

### Group by trait name
In this section we group our data by trait name

```python
newdf= df.groupby('trait name')
newdf.first()
```

### Pick specific trait name
Simply Insert a trait name into the cell below as indicated by the comment

```python
data=newdf.get_group('Fat Taste Perception')
data
```

```python
#drop train name from set for easier analysis
data = data.iloc[: , 1:]
data
```

### Basic distrubution of values based on categories 
In order to group different categories simply remove and insert them in the list below

```python
#group data by occurance of code 
newdata=data.groupby(['assesment', 'ancestry', 'bmi', 'diabetes', 'hypertention']).size()
newdata
```

## Data Visualization
Link to pandas plot documentation
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.html
#### Gender:  0 = Male, 1 = Female
#### Ancestry: 0 = AA, 1 = CAU 
#### BMI: 0 = under, 1 = healthy, 2 = overweight, 3 = obesity 
#### diabetes and hypertention: 1 = yes, 0 = no

```python
#assesment	gender	ancestry	bmi	diabetes	hypertention
# plot the result
newdata.plot(figsize=(25,15), legend=False)
plt.xticks(rotation=30)
plt.show()
```
